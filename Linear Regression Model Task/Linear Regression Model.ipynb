```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('Canada_per_capita_income.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year , df.income , color='red' , marker='x')
```

```python
new_df = df.drop("income" , axis = 1)
```

```python
new_df
```

```python
income = df.income
```

```python
income
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(new_df , income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
