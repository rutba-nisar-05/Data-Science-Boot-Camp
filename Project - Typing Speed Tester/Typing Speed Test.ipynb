```python
import random
import time
import textwrap
import msvcrt
import os

sentences = [
    "In the realm of possibility, where dreams collide with determination and perseverance, lies the potential for extraordinary achievements that transcend the boundaries of ordinary existence." , 
    "Amidst the chaos and cacophony of life, it is in the serene moments of introspection and self-reflection that we discover the depths of our own souls and truly understand the essence of our being." , 
    "Through the intricacies of science and the marvels of technology, humanity has embarked on a relentless pursuit of knowledge, unraveling the mysteries of the universe and pushing the boundaries of what is deemed possible." , 
    "In the tapestry of life, woven with threads of joy and sorrow, love and loss, we navigate the labyrinthine paths, embracing the beauty of each moment and finding solace in the melodies of our shared experiences." , 
    "As the sun sets on the horizon, casting a golden glow upon the world, we are reminded of the fleeting nature of time and the importance of cherishing every precious instant, for life's tapestry is woven with the threads of ephemeral moments." , 
    "In the symphony of existence, each individual plays a unique melody, harmonizing with others to create a cacophony of diversity, where every note, no matter how seemingly insignificant, contributes to the grandeur of the collective human experience."
]

def wpm_calculation(time_duration, sentence):
    words = sentence.split()
    num_words = len(words)
    minutes = time_duration / 60
    wpm = int(num_words / minutes)
    return wpm

def accuracy_calculation(user_input, sentence):
    correct_characters = sum(a == b for a, b in zip(user_input, sentence))
    total_characters = len(sentence)
    accuracy = (correct_characters / total_characters) * 100
    return accuracy

# Randomly select a sentence from the list
r_sentence = random.choice(sentences)

print("\033[1;35m" + "**************************************************** TYPING SPEED TEST ****************************************************")
print('\n')
print("***************************************************************************************************************************")
print('\n')
input("************************************************** PRESS ENTER TO BEGIN ***************************************************")
os.system('cls')

# Provide the sentence that the user needs to type
print("\033[1;35m" + "KINDLY TYPE THE FOLLOWING SENTENCE:")
print("\n")
print(textwrap.fill(r_sentence , width=110))
print("\n")

# Invoke the timer to mark the beginning of the test
start_time = time.time()

# Gather the user's input
user_input = input("\033[1;32m")

# Cease the timer to mark the ending of the test
end_time = time.time()

# Calculate the time taken by user for typing the input
time_duration = end_time - start_time

# Using the function defined previously, calculate the typing speed
typing_speed = wpm_calculation(time_duration, user_input)
accuracy = accuracy_calculation(user_input, r_sentence)

# Present the resultant data to the user
print("\n")
print("\033[1;36m" + "********************** RESULT **********************")
print(f"Duration: {time_duration:.2f} seconds")
print(f"Typing speed: {typing_speed} WPM")
print(f"Accuracy: {accuracy:.2f}%")

user_input = input("\033[32;5mPress any key to exit\033[0m ")
```
